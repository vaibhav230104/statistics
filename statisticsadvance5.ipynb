```python
#Q1
import scipy.stats as stats
z = stats.norm.ppf(0.975)
x = 50
std = 5
n = 100
lower = x - z*std/(n)**0.5
upper = x + z*std/(n)**0.5
print((lower,upper))
```

```python
#Q2
import numpy as np
observed = np.array([30, 40, 50, 20, 15, 45])

expected = np.array([
    0.2 * np.sum(observed), 
    0.2 * np.sum(observed), 
    0.2 * np.sum(observed), 
    0.1 * np.sum(observed), 
    0.1 * np.sum(observed), 
    0.2 * np.sum(observed)
])

chi2, p = stats.chisquare(f_obs=observed, f_exp=expected)

alpha = 0.05
if p < alpha:
    print("The distribution of colors significantly differs from the expected distribution.")
else:
    print("The distribution of colors does not significantly differ from the expected distribution.")
```

```python
#Q3
# Create the contingency table

observed = [[20, 15], [10, 25], [15, 20]]
# Perform the chi-square test
chi2, p,_,_ = stats.chi2_contingency(observed)

# Print the results
print("Chi-square statistic:", chi2)
print("p-value:", p)
```

```python
#Q6
import matplotlib.pyplot as plt
x = np.linspace(0,30,500)
y = stats.chi2.pdf(x,df = 10)
x_fill = np.linspace(15,30,500)
y_fill = stats.chi2.pdf(x_fill,df = 10)
plt.fill_between(x_fill,y_fill,color = 'red',alpha = 0.6, label = 'Area: >15')
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability Density')
plt.title('Chi-square distribution')
plt.plot(x,y,alpha = 0.9,label = 'Chi-square Distribution (df=10)')
plt.legend()
plt.show()
```

```python
#Q8
#Tails= 45 ,Heads = 55
observed = [45,55]
expected = [50,50]
chi2, pvalue = stats.chisquare(observed,expected,1)
alpha = 0.05
critical_value = stats.chi2.ppf(1-alpha,1)
if chi2 > critical_value:
    print('We reject the null hypothesis and coin is unfair')
else:
    print('We accept the null hypothesis and coin is fair')
```
